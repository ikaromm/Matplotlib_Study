```python
import matplotlib.pyplot as plt
import numpy as np

meses = [
    "Jan",
    "Fev",
    "Mar",
    "Abr",
    "Mai",
    "Jun",
    "Jul",
    "Ago",
    "Set",
    "Out",
    "Nov",
    "Dez",
]
vendas = [95, 90, 78, 92, 90, 96, 96, 99, 101, 103, 105, 108]

# usando o np where, crie uma lista de marcadores x para o valor mÃ­nimo e o para o resto
marcadores = np.where(vendas == np.max(vendas), "x", "")
# array(['', '', 'x', '', '', '', '', '', '', '', '', ''], dtype='<U1')

plt.plot(meses, vendas, color = "lime")

for posicao, marcador in enumerate(marcadores):
    if marcador == "x":
        cor = "red"
        tamanho = 80
    else:
        cor = "blue"
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)

#anotaÃ§Ã£o para mostrar o maior valor
plt.annotate(
    "Maior",
    color="red",
    xy=(11, 108),
    xytext=(11.5, 109),
    arrowprops=dict(facecolor="red"),
)

plt.grid(True)
plt.xlabel("MÃªs")
plt.ylabel("Vendas (em milhares)")
plt.title("Crescimento das Vendas ao Longo do Ano")
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

meses = [
    "Jan",
    "Fev",
    "Mar",
    "Abr",
    "Mai",
    "Jun",
    "Jul",
    "Ago",
    "Set",
    "Out",
    "Nov",
    "Dez",
]
vendas = [95, 90, 78, 92, 90, 96, 96, 99, 101, 103, 105, 108]

# cycler para definir as cores
cores = plt.get_cmap("Set2").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

plt.figure(figsize=(8, 4))
plt.plot(meses, vendas, marker="o")

# adicionando um vspan para destacar o intervalo de meses com vendas a partir de julho
plt.axvspan(6, 11, color=cores[1], alpha=0.5)

# adicionando texto para destacar o intervalo de meses com vendas a partir de julho
plt.text(6.1, 89, "RecuperaÃ§Ã£o de crescimento \n (jul/dez)", fontsize=11, color= "slateblue", weight="bold")

plt.xlabel("MÃªs")
plt.ylabel("Vendas (em milhares)")
plt.title("Crescimento das Vendas ao Longo do Ano")

plt.yticks([])
plt.tick_params(axis="x", length=0)
plt.box(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

meses = [
    "Jan",
    "Fev",
    "Mar",
    "Abr",
    "Mai",
    "Jun",
    "Jul",
    "Ago",
    "Set",
    "Out",
    "Nov",
    "Dez",
]
vendas = [95, 90, 78, 92, 90, 96, 96, 99, 101, 103, 105, 108]

cores = plt.get_cmap("tab10").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

plt.figure(figsize=(8, 4))
plt.plot(meses, vendas)

# adicionando um vspan para destacar o intervalo de meses com vendas a partir de julho
plt.axvspan(6, 11, color=cores[1], alpha=0.5)

# adicionando texto para destacar o intervalo de meses com vendas a partir de julho
plt.text(6.1, 89, "RecuperaÃ§Ã£o de crescimento", fontsize=11, color="green", weight="bold")

# usando o np where, crie uma lista de marcadores x para o valor mÃ­nimo e o para o resto
marcadores = np.where(vendas == np.min(vendas), "x", "o")
for posicao, marcador in enumerate(marcadores):
    if marcador == "x":
        cor = "red"
        tamanho = 80
    else:
        cor = cores[0]
        tamanho = 30
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)
    plt.annotate(
        vendas[posicao],
        color=cor,
        xy=(meses[posicao], vendas[posicao]),
        xytext=(-5, 5),
        textcoords="offset points",
    )

plt.ylabel("Vendas (em milhares)")
plt.title("Crescimento das Vendas ao Longo do Ano")

plt.yticks([])
plt.tick_params(axis="x", length=0)
plt.box(False)

plt.show()
```
