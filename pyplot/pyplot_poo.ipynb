# Arrays & Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

cores = plt.get_cmap("Dark2").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

np.random.seed(42)


fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))


# gráfico 1 - linha ------------------------------------------------------------------
months = [
    "Jan",
    "Fev",
    "Mar",
    "Abr",
    "Mai",
    "Jun",
    "Jul",
    "Ago",
    "Set",
    "Out",
    "Nov",
    "Dez",
]
random_numbers_12 = np.random.randint(0, 5, 12)
sales = random_numbers_12

axs[0, 0].plot(months, sales)
axs[0, 0].set_xlabel("Mês")
axs[0, 0].set_ylabel("Vendas (em milhares de reais)")
axs[0, 0].set_title("Crescimento das Vendas ao Longo do Ano")

# gráfico 2 - barras ------------------------------------------------------------------
products = ["Produto A", "Produto B", "Produto C", "Produto D", "Produto E"]
random_numbers_5 = random_numbers_12 = np.random.randint(1, 55, 5)
qnt = random_numbers_5

axs[0, 1].bar(products, qnt)
axs[0, 1].set_xlabel("Produto")
axs[0, 1].set_ylabel("Quantidade de vendas")
axs[0, 1].set_title("Quantidade de vendas por produto")

# gráfico 3 - pizza ------------------------------------------------------------------
category = ["Eletrônicos", "Roupas", "Alimentos", "Livros"]
random_numbers_4 = random_numbers_12 = np.random.randint(0, 15, 4)
sold_qnt = random_numbers_4
highlights = [0.1, 0.05, 0.05, 0.05]  # destacar a primeira fatia

axs[1, 0].pie(sold_qnt, labels=category, explode=highlights, autopct="%.1f")
axs[1, 0].set_title("Categorias de produtos vendidos")

# gráfico 4 - barras empilhadas ------------------------------------------------------
year = [2018, 2019, 2020, 2021]
online_sales = [5, 10, 15, 20]
store_sales = [10, 8, 6, 4]

axs[1, 1].bar(year, online_sales, label="Online")
axs[1, 1].bar(year, store_sales, bottom=online_sales, label="Loja")
axs[1, 1].set_xlabel("Ano")
axs[1, 1].set_ylabel("Vendas (em milhões de reais)")
axs[1, 1].set_title("Vendas online vs loja")
axs[1, 1].set_xticks(year)
axs[1, 1].legend()

# ajustes finais ---------------------------------------------------------------------
fig.suptitle("Evolução de vendas", fontsize=16)
plt.show()
```
