```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

cores = plt.get_cmap("Dark2").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

np.random.seed(42)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
category = ["Eletrnicos", "Roupas", "Alimentos", "Livros"]
random_numbers_4  = np.random.randint(0, 15, 4)
sold_qnt = random_numbers_4
highlights = [0.1, 0.01, 0.01, 0.01]  # destacar a primeira fatia

wedges, texts, autotexts = ax.pie(sold_qnt, labels=category, explode=highlights, autopct= "%.1f", shadow= True)

ax.set_title("Categorias de produtos vendidos")

ax.legend(wedges, category,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
```
