```python
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt
```

```python
from cycler import cycler
# cycler para definir as cores
cores = plt.get_cmap("Set2").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize = (6, 6))

# Plot de barras
axs.bar([0, 1, 2, 3], [10, 15, 13, 12], label="Temperatura")
axs.set_ylabel("Temperatura em Celsius (Cº)", color = cores[0])
#axs.set_xlabel("Ano")
axs.set_xticks([0, 1, 2, 3])
axs.set_xticklabels([2019, 2020, 2021, 2022])

# Plot de linha
ax2 = axs.twinx()
ax2.plot([0, 1, 2, 3], [15, 25, 33, 35], color=cores[1], label="Umidade")  
ax2.set_ylabel("Umidade em porcentagem (%)", color = cores[1])
ax2.set_yticks([0, 5, 10, 15, 20, 25, 30, 35])

# Coleta de alças e rótulos de ambos os eixos
handles, labels = axs.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

# Combina alças e rótulos
combined_handles = handles + handles2
combined_labels = labels + labels2

# Cria uma legenda única
axs.legend(combined_handles, combined_labels, loc = 'upper left')

# Mostrar o gráfico
plt.show()
```
